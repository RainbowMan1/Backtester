```python
import random
from TradingBackend.Backtester import Backtester
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/Nikesh/Desktop/TradingDataset/Bitmex/XBTUSD/1m OHLCV/XBTUSD.csv')
data.set_index('timestamp', inplace = True, drop = True)
data.index = pd.to_datetime(data.index)
data = (data.resample('30min')
               .agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last', 'Volume':'sum'}))
data.sort_index(inplace=True)
data.fillna(method = 'bfill', inplace=True)
```

```python
data['smaller_moving_average'] = data.Close.rolling(10).mean()
data['longer_moving_average'] = data.Close.rolling(20).mean()
```

```python
def entry(self, i):
    if self.data.smaller_moving_average[i]>self.data.longer_moving_average[i] and self.data.smaller_moving_average[i-1]<=self.data.longer_moving_average[i-1]:
        return 1
    else:
        return 0
```

```python
def monitor(self, i):
    if self.data.smaller_moving_average[i]<self.data.longer_moving_average[i] and self.data.smaller_moving_average[i-1]>=self.data.longer_moving_average[i-1]:
        return 1
    else:
        return 0
```

```python
backtester = Backtester(data,lookback=25)
```

```python
backtester.backtest(entry,monitor)
```
